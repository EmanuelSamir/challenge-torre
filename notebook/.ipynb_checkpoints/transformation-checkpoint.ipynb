```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import re
import string
import scipy
```

# Transformation
This notebook is used to transform the database to obtain a cleaner version with the necessary columns to be used for visualization

```python
filename = '../data/reed_uk.csv'
fields = ['category', 'city', 'company_name', 'job_description', 'job_requirements', 'job_title', 'post_date', 'salary_offered']

df = pd.read_csv(filename, usecols = fields)
```

```python
df['job_requirements'] = df['job_requirements'].replace(np.nan, 'No requirements', regex=True)
```

### Add Latitude and Longitude as extra

```python
import requests
import urllib.parse

def get_lat_and_lon(address):
    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'
    response = requests.get(url).json()
    return (response[0]["lon"], response[0]["lat"])
```

```python
cities = df['city'].values
#for city in cities:
cities_unique, counts = np.unique(cities, return_counts=True)

# Consider only the 50 cities with more jobs
cities_unique = [x for y, x in sorted(zip(counts,cities_unique),reverse=True)]
cities_unique = cities_unique[:50]

cities_unique_lon = dict()
cities_unique_lat = dict()
```

```python
it = 0
for city in cities_unique:
    if it%10 == 0:
        print(it)
    cities_unique_lon[city], cities_unique_lat[city] = get_lat_and_lon(city)
    it += 1
```

```python
cities_lon = []
cities_lat = []
for city in cities:
    try:
        cities_lon.append(cities_unique_lon[city])
        cities_lat.append(cities_unique_lat[city])
    except:
        cities_lon.append(np.nan)
        cities_lat.append(np.nan)
        print("Error. Not found " + city)
```

```python
df['lon'] =  cities_lon
df['lat'] =  cities_lat
```

```python
df = df.drop(['job_description'], 1)
```

```python
filename = '../data/reed_uk_cleaner.csv'
csize = 10000
df.to_csv(filename, chunksize=csize)
```
